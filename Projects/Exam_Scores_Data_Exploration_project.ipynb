# 2. Calculating the average score of all three scores
---
Further analysis in this worksheet will be done using the average score of the reading, writing and maths score. Hence this will be calculated first and added as an extra column to the dataframe.

**Calculating the average of all three scores and appending results to the existing dataframe**

```python
def mean_student_scores():
  
 #Create numpy arrays to hold each of the three sets of scores (math, reading, writing)
  math_score=exams["math_score"].to_numpy(np.int64)
  reading_score=exams["reading_score"].to_numpy(np.int64)
  writing_score=exams["writing_score"].to_numpy(np.int64)

  #Creating a new numpy array to hold the average exam score (of the three scores for each student)
  average = (math_score + reading_score + writing_score)//3
  
  #Add the new numpy array as a new column in the dataframe.
  exams['average_scores'] = average.tolist()


  return  exams.head()#average 

mean_student_scores()
```

*Above can be seen that a new column was added ('average_score') containing the mean of all three scores for each pupil.*
