#2. Finding the proportion of each column
---



Because the data set has been previously used elsewhere in this repository looking at the performance of male and female pupils  on maths reading and writing scores the focus this time is first to look at how the data is distributed. Therefore the questions to answer are:


*   The proportion of male and females
*   The proportion of each ethnicity group 


*   The proportion of pupils on standard and free lunches
*   The proportion of the parental education across pupils





```python

def get_proportions():

  male= exams[exams['gender']=='male']['gender'].count()
  female=exams[exams['gender']=='female']['gender'].count()
  ethnicity_A=exams[exams['ethnicity']=='group A']['ethnicity'].count()
  ethnicity_B=exams[exams['ethnicity']=='group B']['ethnicity'].count()
  ethnicity_C=exams[exams['ethnicity']=='group C']['ethnicity'].count()
  ethnicity_D=exams[exams['ethnicity']=='group D']['ethnicity'].count()
  ethnicity_E=exams[exams['ethnicity']=='group E']['ethnicity'].count()
  parental_education_somecollege=exams[exams['parental_education']=='some college']['parental_education'].count()
  parental_education_somehighschool=exams[exams['parental_education']=='some high school']['parental_education'].count()
  parental_education_highschool=exams[exams['parental_education']=='high school']['parental_education'].count()
  parental_education_bachelors=exams[exams['parental_education']=="bachelor's degree"]['parental_education'].count()
  parental_education_masters=exams[exams['parental_education']=="master's degree"]['parental_education'].count()
  parental_education_associates=exams[exams['parental_education']=="associate's degree"]['parental_education'].count()
  
  male_percentage=male*100/1000
  female_percentage=female*100/1000
  ethnicity_A_percentage=ethnicity_A*100/1000
  ethnicity_B_percentage=ethnicity_B*100/1000
  ethnicity_C_percentage=ethnicity_C*100/1000
  ethnicity_D_percentage=ethnicity_D*100/1000
  ethnicity_E_percentage=ethnicity_E*100/1000
 
  parental_ed_somecollege=parental_education_somecollege*100/1000
  parental_ed_somehighschool=parental_education_somehighschool*100/1000
  parental_ed_highschool=parental_education_highschool*100/1000
  parental_ed_bachelors=parental_education_bachelors*100/1000
  parental_ed_masters=parental_education_masters*100/1000
  parental_ed_associates=parental_education_associates*100/1000
  print("male :", male_percentage,"%", "\n female :", female_percentage,"%", "\n ethnicity A :", ethnicity_A_percentage,"%","\n ethnicity B :", ethnicity_B_percentage,"%", "\n ethnicity C :", ethnicity_C_percentage,"%","\n ethnicity D :", ethnicity_D_percentage,"%", "\n ethnicity E :", ethnicity_E_percentage,"%", "\n parent ed. some college:", parental_ed_somecollege,"%","\n parent ed. some high school:", parental_ed_somehighschool, "%", "\n parent ed. high school:",parental_ed_highschool,"%", "\n parent ed. bachelor's:", parental_ed_bachelors,"%", "\n parent ed. master's:", parental_ed_masters,"%","\n parent ed. associate's:", parental_ed_associates,"%",)
  

  plt.title('Ethnicity', fontsize=15)
  data_eth= [8.2,19.7,31.3,25.4, 15.4]
  labels=['group A', 'group B', 'group C', 'group D', 'group E'] 
  colors=sns.color_palette('pastel')
  plt.pie(data_eth, labels = labels, colors=colors, autopct='%1.1f%%')#colors=colors
  plt.show()

  plt.title('Parental Education', fontsize=15)
  data= [24.4,18.5,19.5,11, 7.6, 19.0]
  labels=['some college', 'some high school', 'high school', "bachelor's", "master's", "associate's"] 
  #labels
  colors=sns.color_palette('pastel')
  plt.pie(data, labels = labels,colors=colors, autopct='%1.1f%%')#color=colors
  plt.show()

 
  data= [50.4,49.6]
  labels=['male', 'female'] 
  #labels
  colors=sns.color_palette('pastel')
  plt.title('Gender', fontsize=15)
  plt.pie(data, labels = labels, colors=colors, autopct='%1.1f%%')
  plt.show()
  
  

get_proportions()
```


*   The proportion of male and female in this data set is almost equal(female=49.6%, male=50.4%).
*   The **biggest ethnicity group** was found to be **group C** while group A was found to be the **smallest** (group A=8.2%, group B= 19.7%, group C=31.3%, group D=25.4% and group E=15.4%). 

*   There are **more pupils on standard** than on free lunches (standard = 64.7%, free/reduced=35.3%)
*   The **percentage of pupils whose parent's have a higher education degree** (master's degree=7.6%, bachelor's degree=11 %) **is smaller** compared to the percentage of those pupils whose parents do not have a HE degree (some college=24.4%, high school=19.5%, some high school=18.5%, associate's degree=19.0%)

# Were those pupils scoring higher mostly those completing the course? Were they the ones found to be in ethnicity group E?
---

It has so far been observed that pupils who scored higher

1.   were on standard meals
2.   parents had a master's degree

1.   belong to ethnicity group E
2.   had completed the preparation course

As previously mentioned, students who did the preparation course scored higher than those who did not. We will now look at what proportion of students who did the preparation course and whether these were also those wo scored higher. 






```python
def get_relation(prep):
  plt.figure(figsize=(30,15))
  plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
  plt.subplot(251)
  plt.title('Test Preparation course Vs Gender',fontsize = 15)
  sns.countplot(hue=prep, x="gender", data=exams)

  plt.subplot(254)
  plt.title('Test Preparation course Vs Parental Level Of Education',fontsize = 15)
  sns.countplot(hue=prep, y="parental_education", data=exams)

  plt.subplot(253)
  plt.title('Test Preparation course Vs Lunch',fontsize = 15)
  sns.countplot(hue=prep, x="lunch", data=exams)

  plt.subplot(252)
  plt.title('Test Preparation course Vs Ethnicity',fontsize = 15)
  sns.countplot(hue=prep, y="ethnicity", data=exams)

plt.show()

get_relation("preparation_course")
```

*Results above show that only the minority of students have completed the preparation course. Most of those who completed the course were from ethnicity **group C**, were on standard lunches and  their parents have had 'some college, some high school, high school or an associate's degree'. Comparing with previous results looking at scores it indicates that doing the preparation course is not an indicator of how high a pupil will score in the exams since the higher scores were found in pupils within ethnicity group E (with a low number of stundents completing the course) and whose parents' had master's degrees (also with low numbers of pupils who completed the course).*

Previously we found that those student's with a master's degree had higher scores than their counterparts whose parents had other degrees. It seems that those students with higher scores are only a minority. The majority of those pupils whose parent's had a master's degree are in group B and C. We found that the group scoring higher was group E where parents' with a degree are a minority. In addition, pupils on standard lunches  also seem to have parents' with all sorts of educational background.This analysis seems to suggest that achiving good grades does not depend on parent's educational background but more on ethnicity.*
